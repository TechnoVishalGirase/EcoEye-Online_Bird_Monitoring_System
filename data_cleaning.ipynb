```python
import cv2
import pandas as pd
import os

# Path to the folder containing original images
input_folder = 'C:/Users/Vishal Girase/Documents/COEI/COEI/ORIGINAL'
# C:\Users\Vishal Girase\Documents\COEI\COEI\ORIGINAL

# Path to the folder where you want to save the cut images
output_folder = 'C:/Users/Vishal Girase/Documents/COEI/COEI/Output'

# Path to the summary CSV file
summary_csv_path = 'C:/Users/Vishal Girase/Documents/COEI/COEI/COEI_info.csv'

# Read the summary CSV file
summary_df = pd.read_csv(summary_csv_path)

# Iterate through each row in the summary file
for index, row in summary_df.iterrows():
    image_name = row['File_name']
    width = int(row['Width'])
    height = int(row['Height'])

    # Read the original image
    image_path = os.path.join(input_folder, image_name)
    original_image = cv2.imread(image_path)

    # Ensure the image was successfully loaded
    if original_image is not None:
        # Crop the image based on width and height
        cropped_image = original_image[:height, :width]

        # Save the cropped image to the output folder
        output_path = os.path.join(output_folder, image_name)
        cv2.imwrite(output_path, cropped_image)
        print(f"{image_name} processed and saved.")

    else:
        print(f"Failed to load {image_name}.")

print("Processing complete.")
```
